```python
# Importar librerías necesarias

import numpy as np

import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_squared_error, r2_score

#1.Realizar un análisis exploratorio de los datos para identificar relaciones entre variables, valores atípicos, tendencias, etc.
# lectura de de datos en Python y lo cargamos en la variable Datos

Datos = pd.read_csv('/content/car details v4.csv')
Datos. head(30)

Datos = pd.read_csv('/content/CAR DETAILS FROM CAR DEKHO.csv')
Datos. head(30)

Datos = pd.read_csv('/content/Car details v3.csv')
Datos. head(30)

Datos = pd.read_csv('/content/car data.csv')
Datos. head(30)

# Descripcon de los datos
Datos.describe()


# Identificacion de datos faltantes y atipicos
plt.figure(figsize=(6,3))
sns.boxplot(x=Datos['Selling_Price'])
plt.title('identificación de datos faltantes y atipicos', fontsize=10)

plt.figure(figsize=(6,3))
sns.boxplot(x=Datos['Present_Price'])
plt.title('identificación de datos faltantes y atipicos', fontsize=10)

#2.Preprocesar los datos limpiándolos, tratando valores faltantes y transformándolos según sea necesario.
# Revisión de datos faltantes
print("Cantidad de registros con cero en Car_Name: ", Datos['Car_Name'].isin([0]).sum())
print("Cantidad de registros con cero en Yearl: ", Datos['Year'].isin([0]).sum())
print("Cantidad de registros con cero en Selling_Price: ", Datos['Selling_Price'].isin([0]).sum())
print("Cantidad de registros con cero en Present_Price: ", Datos['Present_Price'].isin([0]).sum())
print("Cantidad de registros con cero en Kms_Driven: ", Datos['Kms_Driven'].isin([0]).sum())
print("Cantidad de registros con cero en Fuel_Type: ", Datos['Fuel_Type'].isin([0]).sum())
print("Cantidad de registros con cero en Seller_Type: ", Datos['Seller_Type'].isin([0]).sum())
print("Cantidad de registros con cero en Transmission: ", Datos['Transmission'].isin([0]).sum())
print("Cantidad de registros con cero en Owner: ", Datos['Owner'].isin([0]).sum())

#Remplazar todos los ceros con el valor medio (promedio) de la columna
Datos['Owner'] = Datos['Owner'].replace(0, Datos['Owner'].mean())

# Revisión nuevamente de los datos faltantes
print("Cantidad de registros con cero en Car_Name: ", Datos['Car_Name'].isin([0]).sum())
print("Cantidad de registros con cero en Yearl: ", Datos['Year'].isin([0]).sum())
print("Cantidad de registros con cero en Selling_Price: ", Datos['Selling_Price'].isin([0]).sum())
print("Cantidad de registros con cero en Present_Price: ", Datos['Present_Price'].isin([0]).sum())
print("Cantidad de registros con cero en Kms_Driven: ", Datos['Kms_Driven'].isin([0]).sum())
print("Cantidad de registros con cero en Fuel_Type: ", Datos['Fuel_Type'].isin([0]).sum())
print("Cantidad de registros con cero en Seller_Type: ", Datos['Seller_Type'].isin([0]).sum())
print("Cantidad de registros con cero en Transmission: ", Datos['Transmission'].isin([0]).sum())
print("Cantidad de registros con cero en Owner: ", Datos['Owner'].isin([0]).sum())

# Identificacion de datos atipicos

plt.figure(figsize=(6,3))
sns.boxplot(x=Datos['Kms_Driven'])
plt.title('identificación de datos faltantes y atipicos', fontsize=10)

#Conteo de datos atipicos

nivel_minimo = 10000
nivel_maximo =500000
Total_atipicos=((Datos['Kms_Driven']<nivel_minimo)|(Datos['Kms_Driven']>nivel_maximo)).sum()
print("el total de datos atipicos: {}". format(Total_atipicos))

# Filtrar Datos atipicos

Datos = Datos[(Datos['Kms_Driven'] >= nivel_minimo) & (Datos['Kms_Driven'] <= nivel_maximo)]

# Validar si se removieron los datos atipicos

Total_atipicos=((Datos['Kms_Driven']<nivel_minimo)|(Datos['Kms_Driven']>nivel_maximo)).sum()
print("el total de datos atipicos: {}". format(Total_atipicos))

#Revisión del tipo de datos y datos faltantes

Datos.info()

#3.Seleccionar las características más relevantes para entrenar el modelo utilizando selección de características.
# Precio del automóvil en función de sus atributos

X = Datos[['Year', 'Present_Price', 'Kms_Driven','Fuel_Type','Fuel_Type', 'Seller_Type', 'Transmission','Owner']]
y = Datos['Selling_Price']

# 4.Dividir el dataset en Train y Test para evaluar correctamente el modelo.
# Separar en entrenamiento y prueba

Datos_codificados = pd.get_dummies(Datos, drop_first=True)
X = Datos_codificados.drop(['Selling_Price'], axis=1)
y = Datos_codificados['Selling_Price']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
print(X_train)
print(X_test)

# 5.Entrenar el modelo configurando los diferentes hiperparámetros.
# crear y entrenar el modelo

from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
modelo.fit(X_train, y_train)

# Predecir
y_pred = modelo.predict(X_test)

X_test.head(10)

# 6.Evaluar el desempeño del modelo en el conjunto de Test con métricas como precisión, recall, F1-score, etc.
# Evaluar
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Evaluar
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R²: {r2:.2f}")

# 7.Realizar las diferentes gráficas que permitan visualizar los resultados del modelo.
import matplotlib.pyplot as plt
import seaborn as sns
#Muestra qué tan cerca están las predicciones de los valores reales. Lo ideal es que los puntos estén cerca de la línea diagonal
plt.figure(figsize=(8,6))
sns.scatterplot(x=y_test, y=y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # línea ideal
plt.xlabel('Valor Real (y_test)')
plt.ylabel('Valor Predicho (y_pred)')
plt.title('Valores Reales vs Predichos')
plt.show()

#Sirve para verificar si los errores están distribuidos normalmente (ideal para modelos lineales)
residuos = y_test - y_pred

plt.figure(figsize=(8,6))
sns.histplot(residuos, bins=30, kde=True)
plt.xlabel('Error (y_test - y_pred)')
plt.title('Distribución de los errores (residuos)')
plt.show()

import pandas as pd

# Extraer coeficientes
coeficientes = pd.DataFrame(modelo.coef_, index=X_train.columns, columns=["Coeficiente"])
coeficientes = coeficientes.sort_values(by="Coeficiente", ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=coeficientes.Coeficiente, y=coeficientes.index)
plt.title("Importancia de las variables (Regresión Lineal)")
plt.xlabel("Peso del coeficiente")
plt.ylabel("Variables")
plt.show()

# 8. Interpretar, analizar y documentar los resultados obtenidos

#Análisis del desempeño del modelo
#El modelo de regresión lineal muestra un coeficiente de determinación R² de 0.88, lo que indica que el modelo explica aproximadamente el 88% de la variabilidad del precio de venta de los autos.

#El RMSE de 1.73 sugiere que el error promedio de predicción es moderado, considerando que los precios de los autos están en un rango aproximado de 0 a 35 unidades monetarias (según tu boxplot de Selling_Price).

#El RMSE de X.XX sugiere que el error promedio de predicción es bajo/moderado/alto, dependiendo del rango de precios.

#Variables más influyentes
#Según el gráfico de coeficientes, las variables con mayor peso positivo en el precio de venta fueron: Present_Price, Year, etc.

#Por otro lado, variables como Owner o algunas categorías de Transmission pueden tener un efecto negativo.

#Visualización de resultados
#El gráfico de Valores Reales vs Predichos muestra una buena alineación con la diagonal, aunque se observan algunas dispersiones para valores extremos.

#La distribución de errores (residuos) es aproximadamente normal, lo cual es favorable para los supuestos del modelo lineal.

#Conclusiones y recomendaciones
#El modelo puede ser útil como una primera estimación del valor de un automóvil en función de sus atributos técnicos.

#Se recomienda probar con modelos más robustos (como Random Forest o XGBoost) para mejorar la precisión, especialmente si se desea mayor rendimiento en valores extremos.

#También se podría mejorar la predicción incluyendo variables como marca, historial de mantenimiento o condiciones del vehículo.
```
